```python
import pandas as pd
movies = pd.read_csv('movies.csv')
data_movie = pd.read_csv('ratings.csv')
data_movie = data_movie.merge(movies, on='movieId', how='left')
data_movie_5 = data_movie.query('rating == 5.0')
data_movie_5.groupby('title').count().sort_values('rating', ascending=False)
```

```python
data_power = pd.read_csv('power.csv')
filtered_countries = data_power[ (data_power['country']=='Lithuania') | (data_power['country']=='Latvia') | (data_power['country']=='Estonia')]
filtered_cat_coun = filtered_countries[ (filtered_countries['category']==4) | (filtered_countries['category']==12) | (filtered_countries['category']==21)]
filtered_cat_coun_qua = filtered_cat_coun[ (filtered_cat_coun['quantity'] >= 0) ]
filtered_cat_coun_qua_year = filtered_cat_coun_qua[ (filtered_cat_coun_qua['year'] <= 2010) & (filtered_cat_coun_qua['year'] >= 2005) ]
data_total = filtered_cat_coun_qua_year.groupby('country').count().filter(items = ['country', 'quantity']).sort_values('quantity', ascending=False)
data_total
# data_power['baltic'] = data_power.country.apply(baltic)
# data_power.baltic.value_counts()
# data_power_baltic = data_power.query('country', Austria)
# data_power_baltic.head()
```
