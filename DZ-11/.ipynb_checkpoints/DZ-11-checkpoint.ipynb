```python
import pandas as pd
data_movie = pd.read_csv('ratings.csv')

def movie_rating(param):
    if param <= 2:
        return 'low'
    elif param <= 4:
        return 'middle'
    elif param <= 5:
        return 'high'
    
data_movie['class'] = data_movie['rating'].apply(movie_rating)

data_movie
```

```python
import pandas as pd
data = pd.read_csv('keywords.csv')
geo_data = {'Центр': ['москва', 'тула', 'ярославль'], 'Северо-Запад': ['петербург', 'псков', 'мурманск'],'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']}
def geo(words):
    for word in words.split():
        for i in geo_data['Центр']:
            if word == i:
                return 'Центр'
        for i in geo_data['Северо-Запад']:
            if word == i:
                return 'Северо-Запад'
        for i in geo_data['Дальний Восток']:
            if word == i:
                return 'Дальний Восток'
            else:
                return 'undefined'

data['region'] = data['keyword'].apply(geo)
# data.tail(50)
data.groupby('region').count()
```
